Question 8:

Split the data into training and testing sets, with 80% training and 20% testing.

IMPORTANT: For grading purposes, please set random_state=42.

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

Question 9:

Use MinMaxScaler to scale the X_train and X_test datasets. Return X_train_scaled, X_test_scaled.

```python
X_train = pd.concat([df_train.iloc[:, :10], df_train.iloc[:,11:]], axis=1).values
len(X_train)
```

```python
y_train = df_train.iloc[:,10].values
len(y_train)
```

```python
X_test = pd.concat([df_test.iloc[:, :10], df_test.iloc[:,11:]], axis=1).values
len(X_test)
```

```python
y_test = df_test.iloc[:,10].values
len(y_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

Question 10:

Build a Logistic Regression model and use that model to make predictions. Return the accuracy up to 2 decimal places (should be around 0.89) and the value of true positive.

```python
from sklearn.linear_model import LogisticRegression
model_lr = LogisticRegression()
model_lr = model_lr.fit(X_train, y_train)
```

```python
y_pred_lr = model_lr.predict(X_test)
y_pred_lr
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
score_lr = accuracy_score(y_pred_lr, y_test)
print(f"The accuracy is {np.round(score_lr, 2)}")
```

```python
from sklearn.metrics import confusion_matrix
matrix_lr = confusion_matrix(y_pred_lr, y_test)
print(matrix_lr)
print(f"The true positive is {matrix_lr[0][0]}")
```

Question 11:

Build a KNN with 3 neighbors model and use that model to make predictions. Return the accuracy up to 2 decimal places (should be around 0.79) and the value of true positive.

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=3)
model_knn = model_knn.fit(X_train, y_train)
```

```python
y_pred_knn = model_knn.predict(X_test)
y_pred_knn
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
score_knn = accuracy_score(y_pred_knn, y_test)
print(f"The accuracy of knn model is {np.round(score_knn,2)}")
```

```python
from sklearn.metrics import confusion_matrix
matrix_knn = confusion_matrix(y_pred_knn, y_test)
print(matrix_knn)
print(f"The true positive is {matrix_knn[0][0]}")
```
